```python
#Carga desde un archivo .xlsx sin indice
df=pd.read_csv("California_EUA.csv",encoding='unicode_escape')
```

## **Etapa 1: Preprocesamientos de datos**

**Valores Nulos**

```python
df.head(2)
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
dfnum = df.select_dtypes(include = ["number"])
dfobj = df.select_dtypes(include = ["object"])
```

```python
valores_nulos=dfnum.isnull().sum()
valores_nulos
```

```python
#Se sustituye los valores por el promedio pues medida de tendencia central más utilizada y nos permite tener un termino medio de los datos
dfnum=dfnum.fillna(round(dfnum.mean(),1))
dfnum
```

```python
valores_nulos=dfnum.isnull().sum()
valores_nulos
```

```python
#Se utilizo un dropna ya que al hacer la sustitución de variables con promedio seguian apareciendo datos, esto es debido a que los datos de estas columnas estaban en ceros.
dfnum= dfnum.dropna(axis=1, how='all')
```

```python
valores_nulos=dfnum.isnull().sum()
valores_nulos
```

```python
valores_nulos=dfobj.isnull().sum()
valores_nulos
```

```python
#Se utilizará el método ffill, dado que sustituye los valores nulos por valores no nulos hacia adelante
dfobj= dfobj.fillna(method="ffill") 
```

```python
valores_nulos=dfobj.isnull().sum()
valores_nulos
```

**Eliminación de outliers con Rango Intercuartílico**


Se utilizo rango intercuantilico ya que te permite encontrar más outliers ya que acota los límites de los datos

```python
y=dfnum

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= dfnum[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
num_iqr= dfnum[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
num_iqr
```

```python
valores_nulos=num_iqr.isnull().sum()
valores_nulos
```

```python
num_iqr=num_iqr.fillna(round(num_iqr.mean(),1))
num_iqr
```

```python
valores_nulos=num_iqr.isnull().sum()
valores_nulos
```

```python
df_limpio= pd.concat([num_iqr, dfobj], axis=1)
df_limpio
```

```python
#Convertir DataFrame a CSV
df_limpio.to_csv("DatosEUA_limpios.csv")
```
