```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl 
```

```python
df = pd.read_csv("DatosMéxico_limpios.csv", low_memory= False)
```

## **Análisis univariado**

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(df)
```

## **A)host_response_time**

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(df['host_response_time'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>110]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('host_response_time')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Tiempo de respuesta del host')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
```

## **C) host_identity_verified**


```python
#Obtengo un análisis univariado de una variable categórica en específico
table2= freq_tbl(df['host_identity_verified'])
table2
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table2[table2['frequency']>1]
Filtro
```

```python
Filtro_index2= Filtro.set_index('host_identity_verified')
Filtro_index2
```

```python
Filtro_index2["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
plt.title("Identidad del host verificada")
```
